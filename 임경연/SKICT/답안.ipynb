```python
from collections import deque


def solution(money, costs):
    # 제일 싼 화폐 단위 확인
    money_order = {0:1, 1:5, 2:10,3:50, 4:100, 5:500}
    cost_value = {}
    for i, j in enumerate(costs):
        cost_value[i] = j * 500 / money_order[i]
    
    cost_value = sorted(cost_value.items(), key = lambda x: x[1])


    answer = 0
    while cost_value:
        moneys = cost_value.pop(0)
        a = moneys[0]

        answer += (money // money_order[a]) * costs[a]

        money = money % money_order[a]
    
    return answer






```
