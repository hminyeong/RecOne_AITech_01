```python
import re
from itertools import combinations

def solution(sentences, n):
    words = []
    s_info = []
    for sentence in sentences:
        Big = re.findall("[A-Z]", sentence)

        score = len(sentence) + len(Big)
        sentence = re.sub(" ", "", sentence)
        word = list(set(sentence.lower()))
        if Big:
            word.append("shift")
        words += word
        word.insert(0, score)
        s_info.append(word)
    words = list(set(words))

    com = list(combinations(words, n))

    answer = 0

    for i in com:
        tmp = 0
        for j in s_info:
            word = j[1:]
            token = 0
            for k in word:
                if k not in i:
                    token = 1
                    break
            if token == 0:
                tmp += j[0]
        answer = max(answer, tmp)

    return answer
```

```python
def solution(arr, brr):
    answer = 0
    for i, s in enumerate(arr):
        if s != brr[i]:
            diff = brr[i] - s
            arr[i + 1] -= diff
            answer += 1

    return answer
```

```python
def solution(abilities, k):
    answer = 0
    if len(abilities) % 2 == 1:
        abilities.append(0)
    abilities.sort()
    rounds = (len(abilities) // 2)

    diffs = []

    for r in range(rounds):
        a = abilities.pop()
        b = abilities.pop()
        diff = a - b

        answer += b
        diffs.append(diff)
    
    diffs.sort()
    for i in range(k):
        answer += diffs.pop()

    return answer
```

```python
from collections import defaultdict

def solution(req_id, req_info):
    answer = []
    sell = []
    buy = []
    record = defaultdict(list)


    for num, info in enumerate(req_info):
        if info[0] == 0:
            buy.append([info[2], -num, info[1], req_id[num]]) # 가격, 등록순서, 수량, 이름
        else:
            sell.append([info[2], num, info[1], req_id[num]])

    sell.sort()
    buy.sort(reverse=True)


    users = list(set(req_id))
    for user in users:
        record[user] = [0, 0]

    while (sell[0][0] <= buy[0][0]) or (len(sell) == 0 or len(buy) == 0):
        sell_info = sell[0]
        buy_info = buy[0]
        amount = min(sell_info[2], buy_info[2])

        record[sell_info[3]][0] -= amount
        record[sell_info[3]][1] += amount * sell_info[0]
        record[buy_info[3]][0] += amount
        record[buy_info[3]][1] -= amount * sell_info[0]


        if amount == sell[0][2] and amount == buy[0][2]:
            sell.pop(0)
            buy.pop(0)
        elif amount == sell[0][2] and amount != buy[0][2]:
            sell.pop(0)
            buy[0][2] -= amount
        else:
            buy.pop(0)
            sell[0][2] -= amount

    for user in record:
        g, s = record[user]
        tmp = user
        if g == 0:
            tmp += " 0"
        elif g < 0:
            tmp += " " + str(g)
        else:
            tmp += " +" + str(g)
        if s == 0:
            tmp += " 0"
        elif s < 0:
            tmp += " " + str(s)
        else:
            tmp += " +" + str(s)
        answer.append(tmp)
    answer.sort()


    

    return answer
```
